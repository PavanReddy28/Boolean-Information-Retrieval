```python
def list_intersection(l1,l2):
    intersection = []
    pointer_l1 = 0
    pointer_l2 = 0
    while(True):
        if (pointer_l1 == len(l1) or pointer_l2 == len(l2)):
            break
        if l1[pointer_l1] == l2[pointer_l2]:
            pointer_l1 += 1
            pointer_l2 += 1
            intersection.append(l1[pointer_l1-1])
        elif l1[pointer_l1] > l2[pointer_l2]:
            pointer_l2 += 1
        else:
            pointer_l1 += 1
    return intersection
```
